```python
import pandas as pd

def get_betting_info(year):
    combined_data = []  # List to collect game data

    for j in range(1, 16):  # Loop through weeks
        game_data = api_instance.get_lines(year=year, week=j)

        for game in game_data:
            game_id = game.id  # Extract the unique identifier for the game
            away_team = game.away_team  # Get the away team name
            away_conference = game.away_conference  # Get the away team's conference
            away_score = game.away_score  # Get the away team's score
            home_team = game.home_team  # Get the home team name
            home_conference = game.home_conference  # Get the home team's conference
            home_score = game.home_score  # Get the home team's score
            season_type = game.season_type  # Get the season type
            start_date = game.start_date  # Get the start date of the game
            
            # Compile all data into a single dictionary for each line
            game_info = {
                    'game_id': game_id,
                    'season': year,
                    'season_type': season_type,
                    'week': j,
                    'start_date': start_date,
                    'away_team': away_team,
                    'away_conference': away_conference,
                    'away_score': away_score,
                    'home_team': home_team,
                    'home_conference': home_conference,
                    'home_score': home_score,
                }
            combined_data.append(game_info)
            i=0
            # Extract betting lines information and flatten it
            for line in game.lines:
                line_info = {
                    'provider{}'.format(i): line.provider,
                    'formatted_spread{}'.format(i): line.formatted_spread,
                    'spread{}'.format(i): line.spread,
                    'over_under{}'.format(i): line.over_under
                }
                i=i+1
                combined_data.append(line_info)  # Append the dictionary to the list

    # Create a DataFrame from the list of dictionaries
    df = pd.DataFrame(combined_data)
    return df
```

```python
bets24=get_betting_info(2024)
bets23=get_betting_info(2023)
bets22=get_betting_info(2022)
bets21=get_betting_info(2021)
bets20=get_betting_info(2020)
```
